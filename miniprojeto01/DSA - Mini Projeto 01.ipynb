#Step 1 - Import Libraries:

```python
# The pandas library works with tabular data
import pandas as pd  

# The numpys library works with math operations
import numpy as np  

# The MatPlotLib works creates graphs
import matplotlib.pyplot as plt  

# The seaborn library show statitics data 
import seaborn as sns  

# The randon library generate randon numbers
import random  

# These datetime and timedelta classes handle dates and time interval
from datetime import datetime, timedelta  

# This MatPlotLib command show graphs
%matplotlib inline  
```

#Step 2 - Generate fictitious sales data:

```python
#We defined a function to generate fictitious data for our mini-project. 
#On this function, we neede generate X sales, with order number, order date, product name, product's category, price, customer ID, cities and state.

def gerar_dados_ficticios(vendas = 600):
    print(f"\nIniciando a geração de {vendas} registros de vendas...")
    produtos = {
        'Laptop Gamer': {'categoria': 'Eletrônicos', 'preco': 6500.00},
        'Mouse Gamer': {'categoria': 'Acessórios', 'preco': 450.00},
        'Teclado Mecânico': {'categoria': 'Acessórios', 'preco': 550.00},
        'Teclado Padrão': {'categoria': 'Acessórios', 'preco': 80.00},
        'Monitor Ultrawide': {'categoria': 'Eletrônicos', 'preco': 2800.00},
        'Cadeira Gamer': {'categoria': 'Móveis', 'preco': 1200.00},
        'Headset 7.1': {'categoria': 'Acessórios', 'preco': 800.00},
        'Placa de Vídeo': {'categoria': 'Hardware', 'preco': 1500.00},
        'SSD 1TB': {'categoria': 'Hardware', 'preco': 600.00}
    }
    lista_produtos = list(produtos.keys())
    cidades_estados = {
        'Cuiabá': 'MT', 'Campo Grande': 'MS', 'Goiânia': 'GO',
        'São Paulo': 'SP', 'Belo Horizonte': 'MG', 'Palmas': 'TO', 'Porto Velho': 'RO'
    }
    lista_cidades = list(cidades_estados.keys())
    dados_vendas = []
    data_inicial = datetime(2026, 1, 1)
    for i in range(vendas):
        produto_nome = random.choice(lista_produtos)
        cidade = random.choice(lista_cidades)
        quantidade = np.random.randint(1, 11)
        data_pedido = data_inicial + timedelta(days = int(i/5), hours = random.randint(0, 23))
        if produto_nome in ['Mouse Gamer', 'Teclado Mecânico', 'Cadeira Gamer']:
            preco_unitario = produtos[produto_nome]['preco'] * np.random.uniform(0.86, 1.0)
        else:
            preco_unitario = produtos[produto_nome]['preco']
        dados_vendas.append({
            'ID_Pedido': 864 + i,
            'Data_Pedido': data_pedido,
            'Nome_Produto': produto_nome,
            'Categoria': produtos[produto_nome]['categoria'],
            'Preco_Unitario': round(preco_unitario, 2),
            'Quantidade': quantidade,
            'ID_Cliente': np.random.randint(50, 150),
            'Cidade': cidade,
            'Estado': cidades_estados[cidade]
        })
    print("Geração de dados concluída.\n")
    return pd.DataFrame(dados_vendas)
```

#Step 3 - Execute the function and explore the data.

```python
vendas = gerar_dados_ficticios(600)
```

#Step 6 - Working with the table

```python
vendas['Faturamento'] = vendas['Preco_Unitario'] * vendas['Quantidade']
```

```python
vendas['Status_Entrega'] = vendas['Estado'].apply(lambda estado: 'Rápida' if estado in ['SP', 'GO', 'MG'] else 'Normal')
```

```python
vendas['Mes'] = vendas['Data_Pedido'].dt.to_period('M')
```

```python
vendas_estado = vendas.groupby('Estado')['Faturamento'].sum().sort_values(ascending = False)
```

```python
faturamento_categoria = vendas.groupby('Categoria')['Faturamento'].sum().sort_values(ascending = False)
```

#Step 7 - Analyzing and creating charts

```python
top_10_produtos = vendas.groupby('Nome_Produto')['Quantidade'].sum().sort_values(ascending = False).head(10)
```

```python
#TOP 10 Best-selling products
sns.set_style("whitegrid")
plt.figure(figsize = (12, 7))
top_10_produtos.sort_values(ascending = True).plot(kind = 'barh', color = 'skyblue')
plt.title('Top 10 Produtos Mais Vendidos', fontsize = 16)
plt.xlabel('Quantidade Vendida', fontsize = 12)
plt.ylabel('Produto', fontsize = 12)
plt.tight_layout()
plt.show()
```

```python
faturamento_mensal = vendas.groupby('Mes')['Faturamento'].sum()
```

```python
faturamento_mensal.index = faturamento_mensal.index.strftime('%Y-%m')
```

```python
faturamento_mensal.map('R$ {:,.2f}'.format)
```

```python
#month by month saler
plt.figure(figsize = (12, 6))

faturamento_mensal.plot(kind = 'line', marker = 'o', linestyle = '-', color = 'green')

plt.title('Evolução do Faturamento Mensal', fontsize = 16)

plt.xlabel('Mês', fontsize = 12)

plt.ylabel('Faturamento (R$)', fontsize = 12)

plt.xticks(rotation = 45)

plt.grid(True, which = 'both', linestyle = '--', linewidth = 0.5)

plt.tight_layout()

plt.show()
```

```python
vendas_estado.map('R$ {:,.2f}'.format)
```

```python
#Sales by state

plt.figure(figsize = (12, 7))

vendas_estado.plot(kind = 'bar', color = sns.color_palette("husl", 7))

plt.title('Faturamento Por Estado', fontsize = 16)

plt.xlabel('Estado', fontsize = 12)

plt.ylabel('Faturamento (R$)', fontsize = 12)

plt.xticks(rotation = 0)

plt.tight_layout()

plt.show()
```

```python
faturamento_categoria.map('R$ {:,.2f}'.format)
```

```python
#Sales by category

from matplotlib.ticker import FuncFormatter

faturamento_ordenado = faturamento_categoria.sort_values(ascending = False)

fig, ax = plt.subplots(figsize = (12, 7))

def formatador_milhares(y, pos):
    """Formata o valor em milhares (K) com o cifrão R$."""
    return f'R$ {y/1000:,.0f}K'

formatter = FuncFormatter(formatador_milhares)

ax.yaxis.set_major_formatter(formatter)

faturamento_ordenado.plot(kind = 'bar', ax = ax, color = sns.color_palette("viridis", len(faturamento_ordenado)))

ax.set_title('Faturamento Por Categoria', fontsize = 16)
ax.set_xlabel('Categoria', fontsize = 12)
ax.set_ylabel('Faturamento', fontsize = 12)

plt.xticks(rotation = 45, ha = 'right')

plt.tight_layout()

plt.show()
```
